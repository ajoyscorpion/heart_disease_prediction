```python
data = pd.read_csv("/Users/ajoy/dataScience/Practice/cancer detection/framingham.csv")
```

```python
data.drop(['education'],inplace=True,axis=1)
```

```python
data.rename(columns={'male':"sex_male"},inplace=True)
```

```python
data
```

```python
data.dropna(axis=0,inplace=True)
```

```python
data
```

```python
# Splitting the Dataset into Test and Train Sets
```

```python
x = np.asarray(data[['age','sex_male','cigsPerDay','totChol','sysBP','glucose']])
```

```python
y = np.asarray(data[['TenYearCHD']])
```

```python
x
```

```python
plt.plot(x)
plt.show()
```

```python
y
```

```python
from sklearn import preprocessing
```

```python
x = preprocessing.StandardScaler().fit(x).transform(x)
```

```python
plt.plot(x)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.8,random_state=24)
```

```python
x_train
```

```python
x_test
```

```python
y_test
```

```python
# Exploratory Data Analysis of Heart Disease Dataset
```

```python
sns.countplot(x='TenYearCHD' , data=data)
plt.show()
```

```python
counts = data['TenYearCHD'].plot()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
```

```python
logreg.fit(x_train,y_train)
```

```python
y_pred = logreg.predict(x_test)
```

```python
y_pred
```

```python
# Accuracy
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy Score : ", accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cf_matrix = confusion_matrix(y_test,y_pred)
```

```python
cf_matrix
```

```python
sns.heatmap(cf_matrix, annot = True, fmt = 'd', cmap = "Greens")
plt.show()
print("Details of confusion matirx : ",classification_report(y_test,y_pred))
```
